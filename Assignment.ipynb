```python
import pandas as pd
import sys
```

```python
url = 'https://docs.google.com/spreadsheets/d/11Js0IzbHBUhA92i47nfmh1ZvyXfDfJUiGEnUbsBvhTM/export?format=csv'
df = pd.read_csv(url,on_bad_lines='skip')
```

```python
df.head(5)
```

```python
# name and position of employees who worked 7 consecutive days
output_file = "output.txt"
with open(output_file, "w") as f:
    # Redirect the stdout to the file
    sys.stdout = f
    df['Time'] = pd.to_datetime(df['Time'])

    consecutive_days = df.groupby('Employee Name')['Time'].apply(lambda x: x.diff().dt.days == 1).groupby(level=0).cumsum()


    employees_7_days = consecutive_days[consecutive_days == 6].index


    for employee in employees_7_days:
      position = df[df['Employee Name'] == employee]['Position ID'].iloc[0]
      print(f"Employee: {employee}, Position: {position}")


    df['Time'] = pd.to_datetime(df['Time'])


    df_sorted = df.sort_values(by=['Employee Name', 'Time'])


    df_sorted['Time Difference'] = df_sorted.groupby('Employee Name')['Time'].diff()


    filtered_employees = df_sorted[(df_sorted['Time Difference'] > pd.Timedelta(hours=1)) &
                               (df_sorted['Time Difference'] < pd.Timedelta(hours=10))]

    print("who have less than 10 hours of time between shifts but greater than 1 hour\n\n")
    for index, row in filtered_employees.iterrows():
      print(f"Employee: {row['Employee Name']}, Position ID: {row['Position ID']}, Position Status: {row['Position Status']}, Time In: {row['Time']}")

    df['Time'] = pd.to_datetime(df['Time'])
    df['Time Out'] = pd.to_datetime(df['Time Out'])


    df['Shift Duration'] = df['Time Out'] - df['Time']

    filtered_employees = df[df['Shift Duration'] > pd.Timedelta(hours=14)]
    print('\n')
    print("Who has worked for more than 14 hours in a single shift\n\n")
    for index, row in filtered_employees.iterrows():
      print(f"Employee: {row['Employee Name']}, Position ID: {row['Position ID']}, Position Status: {row['Position Status']}, Shift Duration: {row['Shift Duration']}")

    sys.stdout = sys.__stdout__
```
